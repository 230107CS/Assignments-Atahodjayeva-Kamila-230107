```python
from PIL import Image
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
img_pil = Image.open("/content/sample_data/Untitled Folder/CAU.png")

print("Image type:", type(img_pil))
print("Image size:", img_pil.size)
print("Image mode:", img_pil.mode)

display(img_pil)
```

```python
img_np = np.array(img_pil)

print("Array shape:", img_np.shape)
print("Data type:", img_np.dtype)
```

```python
img_cv = cv2.cvtColor(img_np, cv2.COLOR_RGB2BGR)

cv2_imshow(img_cv)
```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

img_mpl = mpimg.imread("/content/sample_data/Untitled Folder/CAU.png")

print("Image shape:", img_mpl.shape)

plt.imshow(img_mpl)
plt.axis("off")
```

Q: Why does Matplotlib treat images as NumPy arrays?

Because an image is a matrix of pixel values, and NumPy arrays are efficient for storing and processing numerical data.

```python
from skimage import io, img_as_float
import matplotlib.pyplot as plt


img_sk = io.imread("/content/sample_data/Untitled Folder/CAU.png")

img_float = img_as_float(img_sk)

print("Min pixel value:", img_float.min())
print("Max pixel value:", img_float.max())

plt.imshow(img_float)
plt.axis("off")
```

Task 3.1: Data Type Exploration
Q: Why is using astype(float) discouraged compared to img_as_float? Explain briefly.

Because astype(float) does not rescale pixel values, while img_as_float ensures correct normalization of image intensities.

```python
import cv2
import matplotlib.pyplot as plt

img_gray = cv2.imread("/content/sample_data/Untitled Folder/CAU.png", cv2.IMREAD_GRAYSCALE)

img_bgr = cv2.imread("/content/sample_data/Untitled Folder/CAU.png", cv2.IMREAD_COLOR)

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```

```python
edges = cv2.Canny(img_gray, threshold1=100, threshold2=200)

plt.imshow(edges, cmap='gray')
plt.axis("off")
plt.show()
```

3 What is the main difference between RGB and BGR?

RGB stores colors in the order Red, Green, Blue, while BGR stores them as Blue, Green, Red. OpenCV uses BGR by default, which can cause images to display with swapped colors if not converted.
