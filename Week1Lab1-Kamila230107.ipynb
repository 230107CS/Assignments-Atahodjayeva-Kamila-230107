```python
import cv2
color_image = cv2.imread('/Users/User/Downloads/sunflower.jpg', cv2.COLOR_BGR2RGB)

if color_image is None:
    print("Image not found!")
else:    
    cv2.imshow('color image', color_image)
    print("Image shape (H,W,C):", color_image.shape)

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
grayscale_imagecvt = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)
cv2.imwrite('grayscale_image_cvt.jpg', grayscale_imagecvt)
cv2.imshow('grayscale image cvt', grayscale_imagecvt)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
grayscale_image = cv2.imread('/Users/User/Downloads/sunflower.jpg', cv2.IMREAD_GRAYSCALE)
cv2.imwrite('grayscale_image.jpg', grayscale_image)
cv2.imshow('grayscale image', grayscale_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
_, binary_image = cv2.threshold(grayscale_image, 128, 255, cv2.THRESH_BINARY)
cv2.imshow('binary image', binary_image)
cv2.waitKey(0)
```
