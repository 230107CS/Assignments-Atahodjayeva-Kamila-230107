```python
import cv2 
```

```python
import matplotlib.pyplot as plt
image = cv2.imread('/Users/User/Downloads/Cat.jpg')
image_RGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

subtracted_image = cv2.subtract(image_RGB, 100)

plt.subplot(1,2,1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(subtracted_image)
plt.title('Subtracted Image (-100)')
plt.axis('off')

plt.show()
```

```python
subtracted_image = cv2.subtract(image_RGB, 30)

plt.subplot(1,2,1)
plt.imshow(image_RGB)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(subtracted_image)
plt.title('Subtracted Image (-100)')
plt.axis('off')

plt.show()
```

```python
b, g, r = cv2.split(image)
r = r // 3
result = cv2.merge((b,g,r))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Reduced Red Channel")
plt.axis("off")
plt.show()
```
